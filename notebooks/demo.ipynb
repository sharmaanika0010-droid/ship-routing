## Step 1 — Import Modules

```python
import sys
sys.path.insert(0, '../src')
from routing.optimizer import haversine, RouteOptimizer
from ships.ship import SHIPS
from weather.weather import WeatherData
print('All modules imported successfully!')
```

## Step 2 — Test Distance Calculation

```python
routes = [
    ('Mumbai',   'Mombasa',   19.0, 72.8,  -4.0, 39.7),
    ('Chennai',  'Dubai',     13.0, 80.2,  25.2, 55.2),
    ('Mumbai',   'Singapore', 19.0, 72.8,   1.3,103.8),
    ('Kolkata',  'Perth',     22.5, 88.3, -31.9,115.8),
]
print(f'{'Route':<28} {'Distance':>12}')
print('-' * 42)
for name1, name2, la1, lo1, la2, lo2 in routes:
    d = haversine(la1, lo1, la2, lo2)
    print(f'{name1} → {name2:<18} {d:>8.0f} km')
```

## Step 3 — Ship Speed in Different Weather

```python
import matplotlib.pyplot as plt
weather_conditions = [
    ('Calm',        0,  0.0),
    ('Light wind',  10, 1.0),
    ('Moderate',    20, 2.5),
    ('Rough',       30, 4.0),
    ('Storm',       45, 6.0),
]
for ship_type, ship in SHIPS.items():
    speeds = [ship.effective_speed(w, h) for _, w, h in weather_conditions]
    labels = [c[0] for c in weather_conditions]
    plt.plot(labels, speeds, marker='o', label=ship_type.capitalize())
plt.title('Ship Speed vs Weather Conditions')
plt.xlabel('Weather'); plt.ylabel('Speed (knots)')
plt.legend(); plt.grid(True, alpha=0.3); plt.tight_layout()
plt.savefig('../docs/speed_weather.png', dpi=100)
plt.show()
print('Chart saved to docs/speed_weather.png')
```

## Step 4 — Run Route Optimizer

```python
import time
ship    = SHIPS['cargo']
weather = WeatherData()
opt = RouteOptimizer(
    start=(19.0, 72.8),
    end=(-4.0, 39.7),
    ship=ship,
    weather_client=weather,
    mode='time'
)
print('Calculating Mumbai → Mombasa route...')
t0 = time.time()
path, cost = opt.calculate()
print(f'\nTime taken : {time.time()-t0:.1f} seconds')
print(f'Waypoints  : {len(path)}')
print(f'Route cost : {cost:.2f} hours')
```

## Step 5 — Compare Time vs Safety Mode

```python
results = {}
for mode in ['time', 'safety']:
    opt = RouteOptimizer((19.0,72.8), (-4.0,39.7), SHIPS['cargo'], WeatherData(), mode=mode)
    path, cost = opt.calculate()
    results[mode] = {'waypoints': len(path), 'cost': cost}

print('\nMode Comparison — Mumbai → Mombasa (Cargo Ship)')
print('='*45)
for mode, r in results.items():
    print(f'  {mode.upper():<8} | Waypoints: {r["waypoints"]:3d} | Cost: {r["cost"]:.2f}')
```
